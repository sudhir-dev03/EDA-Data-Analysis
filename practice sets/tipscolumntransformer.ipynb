```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder 

from sklearn.preprocessing import OrdinalEncoder 
```

```python
x = df.drop(columns = ['smoker'])
y = df['smoker']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train
```

```python
df = df.drop(columns = ['size'])
```

```python
# simpleimputer for numerical data 
si = SimpleImputer(strategy = 'mean')
x_train_tip = si.fit_transform(x_train[['tip']]) 
x_test_tip = si.transform(x_test[['tip']])
x_train_tip.shape
```

```python
#ordinal encoding for categorical data
oe = OrdinalEncoder(categories=[['Male' , 'Female']])
x_train_sex = oe.fit_transform(x_train[['sex']])
x_test_sex = oe.transform(x_test[['sex']])
x_train_sex.shape
```

```python
#onehotencoder because day and time has morethen two subcategories
ohe = OneHotEncoder(drop = 'first', sparse_output=False)
x_train_daytime = ohe.fit_transform(x_train[['day', 'time']])
x_test_daytime = ohe.fit_transform(x_test[['day', 'time']])
x_train_daytime.shape
```

```python
x_train_total_bill = x_train.drop(columns = ['tip', 'sex', 'day', 'time']).values

x_test_total_bill = x_test.drop(columns = ['tip', 'sex', 'day', 'time']).values
```
