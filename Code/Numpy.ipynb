```python
# View Vs Copy

# View means modification in original data and copy means changes in duplicate data will not reflect
# the original data .
```

```python
a = np.array([10,20,30,40,50,60,70,80])
a
```

```python
a[3:6] = 0
```

```python
a
```

```python
a = np.array([10,20,30,40,50,60,70,80])
b = a[3:6].copy()
b[:] = 0
b
```

```python
a
```
