```python
import plotly.graph_objects as go 
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
import pandas as pd
```

```python
df = pd.read_csv('c:\\Users\\SUDHIR FAUJDAR\\Desktop\\Adv python\\EDA-Data-Analysis\\files\\Attrition - Attrition.csv')
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
# check the percentage 
attr_df = df[df['Attrition'] == 'Yes']
```

```python
attr_df
```

```python
attrition_by_dpt = attr_df.groupby(['Department']).size().reset_index(name = "count")
```

```python
attrition_by_dpt
```

```python
# create a donut chart
fig = go.Figure(data=[go.Pie(
    labels = attrition_by_dpt['Department'],
    values = attrition_by_dpt['count'],
    hole = 0.4,
    marker = dict(colors=['red' , 'green']),
    textposition = 'inside'
)])
```

```python
# upadate the layout
fig.update_layout(title = 'Attrition by Department')
# show the chart
fig.show()
```

```python
attrition_by_edf = attr_df.groupby(['EducationField']).size().reset_index(name = "count")
```

```python
attrition_by_edf
```

```python
fig = go.Figure(data=[go.Pie(
    labels = attrition_by_edf['EducationField'],
    values = attrition_by_edf['count'],
    hole = 0.4,
    marker = dict(colors=['red' , 'green']),
    textposition = 'inside'
)])
```

```python
fig.update_layout(title = 'Attrition by Education Field')

fig.show()
```
