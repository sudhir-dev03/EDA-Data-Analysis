```python
# encoding -- get_dummies
# column --> has_covid['yes', 'no'] ----> when we apply get_dummies, we will have two columns: has_covid_yes and has_covid_no
# we can drop one of the columns to avoid multicollinearity, for example, we can drop has_covid_no, and keep has_covid_yes, which will indicate whether the person has covid or not.    
```
