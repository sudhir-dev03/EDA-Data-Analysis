```python
df = df.drop(columns = ['user_id','age','country','income_bracket','account_created_date','preferred_content_type','usage_purpose','daily_notifications']) 
```

```python
df.head(5)
```

```python
df = df.drop(columns = ['occupation','education_level'])
```

```python
df.shape
```

```python
a = pd.get_dummies(df, drop_first=True)
```

```python
a = a.astype(int)
```

```python
a.head(5)
```

```python
# apply standardization
```

```python
x = a.drop(columns = ['comments_per_day'])
x
```

```python
y = a['comments_per_day']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train) 
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```

```python
# applying minmax scaler
```

```python
df = pd.read_csv('../files/social_media_user_behavior.csv')
```

```python
df.head()
```

```python
# average age 
avg_age = df['age'].mean()
avg_age
```

```python
#  most common gender
most_common_gender = df['gender'].mode()[0]
most_common_gender
```

```python
# top countries 
top_countries = df['country'].value_counts().head()
top_countries
```

```python
#  most used platform 
most_used_platform = df['primary_platform'].mode()[0]
most_used_platform
```

```python
# task 4 --- user behavior analysis 
```

```python
# analyse screen time
df['daily_screen_time_minutes'].describe()
```

```python
# addiction level --- in respect to addiction level used by users.
df.groupby('addiction_level_1_to_10')['daily_screen_time_minutes'].mean()
```

```python
# sleep hours per night --- in respect to sleep hours used by users.
df.groupby('sleep_hours_per_night')['daily_screen_time_minutes'].sum()
```
