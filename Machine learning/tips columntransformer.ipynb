```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import OrdinalEncoder 
```

```python
df = pd.read_csv('../files/tips - tips.csv')
```

```python
df.head()
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns = 'tip'),df['tip'],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers = [
    ('t1', SimpleImputer(), ['total_bill']),
    ('t2', OneHotEncoder(sparse_output=False, drop='first'), ['sex']),
    ('t3', OrdinalEncoder(categories=[['Lunch', 'Dinner']]), ['time'])
], remainder='passthrough')
```

```python
transformer.fit_transform(x_train).shape
```
